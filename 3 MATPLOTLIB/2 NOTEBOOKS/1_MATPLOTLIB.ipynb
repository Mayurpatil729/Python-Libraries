```python
# BASIC
import matplotlib.pyplot as plt     
%matplotlib inline
x=[1,2,3,4,5,6,7] 
y=[50,51,52,48,47,49,46]
# plt.plot(x, y,'rD--')
# plt.plot(x, y,color='red',marker='d',lw="2",ls="dashdot")
plt.plot(x, y,color="#900000",lw="2",ls="dashdot",marker="+",markersize=10,alpha=0.5)
plt.xlabel("DAY")
plt.ylabel("TEMPERATURE")
plt.title("WEATHER OF mumbai",color="red",size="15")
-
```

```python
# Axes labels,legend,Grid
import matplotlib.pyplot as plt
%matplotlib inline
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]

plt.xlabel("DAYS")
plt.ylabel("Temperature")
plt.title("Weather",color='red',size='15')

plt.plot(days,max_t,label='Max')
plt.plot(days,min_t,label='Min')
plt.plot(days,avg_t,label='Avg')

plt.grid()

# plt.legend(loc="lower left")
# plt.legend(loc="upper right")
plt.legend(loc=0,shadow=True,fontsize='large')
```

```python
#BARCHART
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]

xpos=np.arange(len(company))
xpos

plt.ylabel("Revenue(Bin)")
plt.title("US Tech Stocks")

plt.xticks(xpos,company)
plt.bar(xpos-0.2,revenue,width=0.4,label="Revenue",color="red")
plt.bar(xpos+0.2,profit,width=0.4,label="Profit",color="green")
plt.legend()
```

```python
plt.ylabel("Revenue(Bin)")
plt.title("US Tech Stocks")

plt.yticks(xpos, company)
plt.barh(xpos-0.2, revenue,  label="Revenue", color="red")
plt.barh(xpos+0.2, profit, label="Profit", color="green")
plt.legend()
```

```python
#HISTOGRAMS
import matplotlib.pyplot as plt
%matplotlib inline
blood_sugar=[113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,color='g')
# plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,color='g',histtype='step')
```

```python
blood_sugar_men= [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women=[67,98,89,120,133,150,84,69,89,79,120,112,100]
plt.hist([blood_sugar_men, blood_sugar_women], bins=[
         80, 100, 125, 150], rwidth=0.95, color=['green','orange'],label=['men','women'],orientation="horizontal")
plt.legend()

plt.xlabel("Sugar range")
plt.ylabel("Total no of patients")
plt.title("Blood sugar anaylis")
```
